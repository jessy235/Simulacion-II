Caminatas aletorias en 1D


```python
import random
import matplotlib.pyplot as plt
```

```python
#pasos
n_pasos = 1000

#posiciones
y = [0] # punto inicial
x = [0]
```

```python
#generar caminata
for i in range(1,n_pasos + 1):
   paso = random.choice([-1,1]) # elegir subir o bajar
   nuevo_y = y[-1] + paso
   y.append(nuevo_y)
   x.append(i)
```

```python
#grafica
plt.figure(figsize=(10,5))
plt.plot(x, y, marker='o', markersize=3, linestyle='-', linewidth=1)
plt.title("Caminata aleatoria (partiendo desde y=0)")
plt.xlabel("numero de pasos")
plt.ylabel("posicion (y)")
plt.axhline(0, color='red', linestyle = '--') #linea base en y
plt.grid(True, linestyle=':')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
```

```python
# Número de pasos
n = 1000

# Posición inicial
x, y = [0], [0]
```

```python
# Generamos los pasos
for i in range(n):
    dx = random.choice([-1, 1, 0, 0])  # Movimiento en X
    dy = random.choice([0, 0, -1, 1])  # Movimiento en Y
    x.append(x[-1] + dx)
    y.append(y[-1] + dy)
```

```python
# Graficamos la caminata
plt.plot(x, y, marker='o', markersize=3)
plt.title("Caminata Aleatoria 2D")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.axis('equal')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Posición inicial
x, y, z = [0], [0], [0]
```

```python
# Generar pasos aleatorios
for i in range(n):
    dx = random.choice([-1, 0, 1])
    dy = random.choice([-1, 0, 1])
    dz = random.choice([-1, 0, 1])
    x.append(x[-1] + dx)
    y.append(y[-1] + dy)
    z.append(z[-1] + dz)
```

```python
# Graficar la caminata en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, marker='o', markersize=3)
ax.set_title("Caminata Aleatoria 3D")
plt.show()
```
